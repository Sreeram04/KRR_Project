```python
def userful_count(data):
    grouped = data.groupby(['condition']).size().reset_index(name='user_size')
    data = pd.merge(data,grouped,on='condition',how='left')
    return data


file3 =  userful_count(file3) 
file3.head()
```

```python
file3['usefulCount'] = file3['usefulCount']/file3['user_size']
```

```python
file3.head()
```

```python
file3['dl_pred'] = file2['dl_pred']
file3.head()
```

```python
file3['total_pred'] = (file3['ml_pred']+file3['dl_pred'])*file3["usefulCount"]
file3.head()
```

```python
temp = file3.groupby(['condition','drugName']).agg({'total_pred' : ['mean']})
temp
```
